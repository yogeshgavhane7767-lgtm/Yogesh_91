write a program ask the user take two number print subtraction multiplication and division

```python
a= int(input("Enter your first number :"))
b= int(input("Enter you second number :"))
subtraction = a - b
multiplication = a * b
div = a / b
print(f"The subtraction of {a} and {b} is {subtraction}")
print(f"The multiplication of {a} and {b} is {multiplication}")
print(f"The division of {a} and {b} is {div}")
```
