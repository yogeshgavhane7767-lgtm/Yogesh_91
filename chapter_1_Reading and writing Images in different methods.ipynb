```python
# import the opencv library 
import cv2 
vid = cv2.VideoCapture(0) 
  
while True:    
    ret, frame = vid.read()
    cv2.imshow('frame', frame)   
    if cv2.waitKey(1) & 0xFF == ord('q'): 
        break
vid.release() 
cv2.destroyAllWindows()
```

```python
import cv2 
vid = cv2.VideoCapture("los_angeles.mp4") 
  
while True:    
    ret, frame = vid.read() 
    if not ret:
        break
    cv2.imshow('frame', frame)   
    if cv2.waitKey(1) & 0xFF == ord('q'): 
        break
vid.release() 
cv2.destroyAllWindows()
```
